```python
def gradient_descent(x, y, learning_rate=0.01, iterations=1000):
    m = len(x)  # Number of data points
    theta_0, theta_1 = 0, 0  # Initialize parameters to zero 
    for _ in range(1000):
        # Compute predictions
        y_pred = theta_1 * x + theta_0
        # Compute gradients
        d_theta_0 = (1/m) * np.sum(y_pred - y)
        d_theta_1 = (1/m) * np.sum((y_pred - y) * x)
        # Update parameters
        theta_0 -= learning_rate * d_theta_0
        theta_1 -= learning_rate * d_theta_1
    return theta_0, theta_1
theta_0, theta_1 = gradient_descent(x, y)
print(f"Optimized theta_0 (Intercept): {theta_0:.4f}")
print(f"Optimized theta_1 (Slope): {theta_1:.4f}")
```

```python
X_line = x
Y_line = theta_1 * x + theta_0  # Compute Y values
# Plot the data points
plt.scatter(x, y, color='blue', label='Data Points')
# Plot the regression line
plt.plot(X_line, Y_line, color='red', label=f'Best-fit Line: y = {theta_1:.2f}x + {theta_0:.2f}')
# Labels and title
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Linear Regression Line')
plt.legend()
plt.show()
```

```python
import numpy as np

def gradient_descent(x, y, learning_rate=0.01, iterations=1000):
    m = len(y)  # Number of data points
    theta_0, theta_1, theta_2 = 0, 0, 0  # Initialize parameters to zero

    for _ in range(iterations):
        # Compute predictions
        y_pred = theta_1 * x[:, 0] + theta_2 * x[:, 1] + theta_0

        # Compute gradients
        d_theta_0 = (1/m) * np.sum(y_pred - y)
        d_theta_1 = (1/m) * np.sum((y_pred - y) * x[:, 0])
        d_theta_2 = (1/m) * np.sum((y_pred - y) * x[:, 1])

        # Update parameters
        theta_0 -= learning_rate * d_theta_0
        theta_1 -= learning_rate * d_theta_1
        theta_2 -= learning_rate * d_theta_2

    return theta_0, theta_1, theta_2

# Example dataset
x = np.array([[1, 4], 
              [2, 5], 
              [3, 8],
              [4, 2]])

y = np.array([1, 6, 8, 12])  

theta_0, theta_1, theta_2 = gradient_descent(x, y)

print(f"Optimized theta_0 (Intercept): {theta_0:.4f}")
print(f"Optimized theta_1 (Slope): {theta_1:.4f}")
print(f"Optimized theta_2 (Slope): {theta_2:.4f}")
```
